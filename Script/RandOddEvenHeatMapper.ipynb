# RandOddEvenHeatMapper 

First, neccessary libraries should be imported.

```python
import numpy as np
import matplotlib
from matplotlib.pylab import *
import matplotlib.pyplot as plt
import warnings
```

Now, let's define the size and range of elements of the random matrix! Both can be changed by editing the cell. 

```python
required_size=100   #please state the size you require
number_range=100    #please state the number range you require
```

After deciding the size and range of the matrix, we can create a random matrix.

```python
random_matrix=np.random.randint(number_range+1, size=(required_size, required_size))
```

If you want to see the matrix produced according to your inputs, you can run the cell below or skip it!

```python
np.set_printoptions(edgeitems=required_size) #to see the all items of the matrix
random_matrix
```

Let's create a heatmap from the matrix to see the context. 

```python
#Creates a heatmap figure
fig, ax = plt.subplots(figsize=(14,12))
im = ax.imshow(random_matrix, cmap=plt.cm.PuOr, origin='lower')      
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.tick_params(labelsize=24)
for tick in cbar.ax.get_yticklabels():
    tick.set_fontname("Times New Roman")
                      
#Assign the title and change font size and font names.    
ax.set_title("Random Matrix", fontname= "Times New Roman", fontsize=42, pad=10)
ax.tick_params(labelsize=24)
for tick in ax.get_xticklabels():
    tick.set_fontname("Times New Roman")
for tick in ax.get_yticklabels():
    tick.set_fontname("Times New Roman")
             
fig.tight_layout()
                      
#Save the figure.
plt.savefig('RandMatrixHeatMap.jpg', bbox_inches='tight') 
```

Now let's check all the elements in the matrix if they are odd or even!

```python
OddEvenMatrix=random_matrix%2==0
#OddEvenMatrix #to print out the matrix please remove the '#' sign at the beginning of the line.
```

Turn logical elements to 0 and 1 which 1 stands for even and 0 stands for odd numbers!

```python
OddEvenMatrix=OddEvenMatrix.astype(int)
#OddEvenMatrix #to print out the matrix please remove the '#' sign at the beginning of the line.
```

```python
#Creates a heatmap figure
fig, ax = plt.subplots(figsize=(14,12))
im = ax.imshow(OddEvenMatrix, cmap=plt.cm.BuPu, origin='lower')      
cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.tick_params(labelsize=24)
for tick in cbar.ax.get_yticklabels():
    tick.set_fontname("Times New Roman")
    
#change the color bar labels to Odd and Even.
cbar.ax.set_yticklabels(["Odd", "", "", "", "", "Even"]) 
warnings.filterwarnings("ignore") #ignores the warnings

#Assign the title and change font size and font names.
ax.set_title("Random Odd-Even Matrix", fontname= "Times New Roman", fontsize=42, pad=10)
ax.tick_params(labelsize=24)
for tick in ax.get_xticklabels():
    tick.set_fontname("Times New Roman")
for tick in ax.get_yticklabels():
    tick.set_fontname("Times New Roman")
    
fig.tight_layout()


#Save the figure.
plt.savefig('RandOddEvenMatrixHeatMap.jpg', bbox_inches='tight') 
```
